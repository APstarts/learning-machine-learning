```python
import numpy as np
import random
```

```python
points = [
    (1, 1),
    (2, 1),
    (1, 2),
    (10, 10),
    (11, 11),
    (10, 11)
]
```

```python
def distance(a,b):
    return np.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)
```

```python
def initalize_centroids(points,k):
    centroids = random.sample(points,k)
    return centroids
```

```python
centroids = initalize_centroids(points,2)
distances = []
for p in points:
    for c in centroids:
        dist = distance(p,c)
        distances.append(dist)
print(distances.index(min(distances)))
```

```python
def assign_points(points,centroids):
    clusters = [[] for _ in centroids]

    for p in points:
        distances = []
        for c in centroids:
            temp_distance_holder = distance(p,c)
            distances.append(temp_distance_holder)
        closest_index = distances.index(min(distances))
        clusters[closest_index].append(p)
    return clusters
```

```python
def recompute_centroids(clusters):
    new_centroids = []
    for cluster in clusters:

        if len(cluster) == 0:
            continue

        x = sum(p[0] for p in cluster)/len(cluster)
        y = sum(p[1] for p in cluster)/len(cluster)

        new_centroids.append((x,y))
    return new_centroids
```

```python
def kmeans(points,k, max_iters=100):
    centroids = initalize_centroids(points,k)

    for _ in range(max_iters):
        clusters = assign_points(points,centroids)
        new_centroids = recompute_centroids(clusters)

        if new_centroids == centroids:
            break
        centroids = new_centroids
    return centroids, clusters
```

```python
final_centroids, final_clusters = kmeans(points, k=2)

print("Centroids:", final_centroids)
print("Clusters:", final_clusters)
```
